This project is a digital inquiry of an enterprise. It enables customers make inquiry of the enterprise's services and its availability at the comfort of their homes.

**Anas Villa.**

Anas Villa is into hospitality business. They have world eatery, they have high standard hotel, and they are also into party rentals. In addition, they have air-conditioned conference hall for events.

```python
def service_options():
    
    Guest_House = {1: "Room Booking", 2: "Check Availability", 3: "Facilities", 4: "Go back to menu"}
    
    while True:
        House_option = input("Select the guest house option you want: ")
    
        try:
            House_option = int(House_option)
            if House_option > 0:
                if House_option in [1, 2, 3]:
                    return Guest_House[House_option]
                elif House_option == 4:
                    return Guest_House
                else:
                    print("Invalid Input")
            else:
                print("Entry must be a positive integer")
        except ValueError:
            print("Invalid Entry. Enter a corresponding number between 1 and 4")
```

```python
def inquiry():
    while True:
        try:
            Inquiry_input = int(input("Which of our service do you want? "))
            
            if 1 <= Inquiry_input <= 3:
                return Inquiry_input
            elif Inquiry_input == 4:
                return None
            else:
                print("Wrong Input! Your entry must be a number between 1 and 4")
        except ValueError:
            print("Invalid Input! Your entry must be a positive integer")
            

# Main Program            
while True:
    display_services_offered()
    user_input = inquiry() 
    
    if user_input == 1:
        guest_house_option = service_options()  
        display_guest_house_options(guest_house_option)  
        continue
    elif user_input == 2:
        display_eatery_options()
        continue
    elif user_input == 3:
        rentals()    
        continue
    elif user_input is None:
        print("Exiting. Thank you for choosing Anas Villa.")
        break
    else:
        print("Invalid choice. Please enter a number between 1 and 4.")
```
